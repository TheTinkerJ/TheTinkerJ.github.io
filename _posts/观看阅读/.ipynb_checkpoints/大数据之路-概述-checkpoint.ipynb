# 大数据之路读书笔记-概述

## 阿里巴巴大数据体系架构

自下而上一共四层
- 数据采集
- 数据计算
- 数据服务
- 数据应用

## 关于数据采集层

主要采集两方面的用户行为数据:
- web端:Aplus.JS
- APP端:UserTrack

除了数据埋点的设计,还有高性能高可靠的数据传输体系(TimeTunnel):
1) 数据库的增量数据传输
2) 日志数据传输

TT作为`数据传输`服务的基础架构,支持`实时流式计算`和`时间窗口的批量计算`

数据采集层完成数据的生产和传输任务.

## 关于数据计算层

阿里的数据计算层包括两大体系:
1. 数据存储与计算云平台(MaxCompute离线计算平台以及六十计算平台StreamCompute)  
2. 数据整合及管理体系(OneData)

其中
- `MaxCompute`离线大数据平台,强大的存储计算能力
- `StreamCompute`支持流式计算需求
- `OneData`是数据整合及管理的方法体系和工具

关于计算频率
- 阿里的离线数据仓库的计算频率 `T-1`每天凌晨处理上一天的数据
- 双十一时实时数据直播大屏就是实时仓库的应用

关于阿里的数据加工链路:
- 数据层 ODS(Operational Data Store)
- 明细数据层DWD(Data Warehouse Detail)
- 汇总数据层DWS(Data Warehouse Summary)
- 应用层数据ADS(Application Data Store)

关于阿里的元数据
> 元数据: 主要面对数据发现,数据管理
> - 元数据整合数据平台各个系统的元信息，例如，分布式存储的路径、存储量、变更、新增、删除等等
> - 元数据是数据的数据，在数据平台的中原信息的收集、汇总和传递将数据平台各个模块整合起来。元数据管理系统是收集线上db、solor集群、hbase集群和hive集群的元数据信息，并将其传递给各其他系统。
> - 元数据可以帮助数据平台解决“有哪些数据”、“数据存储有多少”、“数据流中的血缘关系”、“如何找到我需要的数据”、“如何使用数据”和“数据的生产进度”，并对上游线上业务库的同步进度和元信息变更进行监控，及时进行影响分析及预警。
> - 元数据管理系统为大数据平台绘制数据地图、统一数据口径、标明数据方位、分析数据关系、管理模型变更及精确到字段级别的影响分析。打通上下游数据继承关系断层，为数据质量维护和业务逻辑可视化打下坚实基础。

- 据源元数据、
- 数据仓库元数据
- 数据链路元数据
- 工具类元数据
- 数据质量类元数据

## 关于数据服务层

整合和计算好后的数据将会被存放在多种数据数据库之上:
- MySQL
- HBase
- ApsaraDB (阿里云数据库)
- 表格存储

数据服务的目的是使应用对底层数据存储透明,访问方便高效

数据服务层对外提供统一服务的平台是数据服务平台`OneService`

### OneService

- 数据源: 数据仓库整合计算好的数据
- 服务方式: 通过对外提供接口的方式提供数据服务
- 服务内容:简单数据查询服务,复杂数查询服务,实施数据推送三个主要业务
