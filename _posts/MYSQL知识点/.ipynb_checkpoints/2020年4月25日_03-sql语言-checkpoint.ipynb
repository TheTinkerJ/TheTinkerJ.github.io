## 基本的SQL查询结构

```SQL
SELECT [DISTINCT] target-list
FROM relation-list
WHERE qualification
```

- target-list 查询结果
- relation-list 从那些表里查
- qualification 查询结果应该满足的条件

与关系演算的关系
$$
\{select\ target-list | (From\ relation-list)\  \wedge (WHERE\ qualification)\}
$$

SQL查询的基本执行策略:
- 根据`FROM`中的`relation-list`进行笛卡尔乘积
- 利用`WHERE`中的 条件`qualification`进行筛选
- 根据`SELECT`中的`target-list`做一次投影\[以及对结果的排序,去重复\]

### 例1

```SQL
SELECT S.sname
FROM Sailors S,Reserves R
WHERE S.sid = R.sid AND R.bid = 103
```

该例的概念执行流程
- S,R先进行笛卡尔乘积
- 进行筛选
- 投影姓名

````python
exec_sql( """
SELECT S1.sname
FROM S1,R1
WHERE S1.sid = R1.sid
UNION
SELECT S2.sname
FROM S2,R1
WHERE S2.sid = R1.sid
""")
````

嵌套查询:
- 非关联嵌套查询
- 关联嵌套查询

````python
# 非关联嵌套查询-找到订过103的水手
exec_sql("""
SELECT S1.sname 
FROM S1
WHERE S1.sid in (
    SELECT R.sid
    FROM R1 as R
    WHERE R.bid=103
    )
""")
````

````python
# 关联嵌套查询-找到订过103的水手
exec_sql("""
SELECT S.sname
FROM S1 AS S
WHERE EXISTS (
    SELECT * 
    FROM R1 AS R
    WHERE R.bid=103 AND S.sid=R.sid 
    )
""")
````

````python
# 预定过103且只有一个水手预定过的船
exec_sql("""
SELECT bid
FROM R1 AS R
WHERE bid NOT IN(
    SELECT bid 
    FROM R1 AS R2 
    WHERE R2.sid!=R.sid 
    )
""")
````

## SQL中的除法

- 找到订购过所有船的水手
- 找到一个水手,这个水手不可能存在没订过某一种船

```SQL
SELECT S.sname 
FROM Sailors S
- 不存在满足下述条件元素
WHERE NOT EXISTS( 
    - 选出所有船
    (SELECT B.bid
     FROM Boats B)
     EXCEPT
    - 减去水手曾经预过的船
    (SELECT R.bid
     FROM Reserves R
     WHERE R.sid=S.sid) 
)
```

不使用`EXCEPT`的方法

```SQL
SELECT S.sname
FROM Sailors S
- 不存在这样的船
WHERE NOT EXISTS(
    SELECT B.bid
    FROM Boats B
    - 什么样的船,没有订购过的船
    WHERE NOT EXISTS(
        SELECT R.bid 
        FROM Reserves R
        WHERE R.bid=B.bid AND R.sid=S.sid
    )
)
```

## GROUP方法,分组运算

需求:找到每个级别中最年轻的水手

引人分组计算之后的SQL基本框架
```SQL
SELECT [DISTINCT] target-list
FROM relation-list
WHERE qualification
GROUP BY grouping-list
HAVING grouping-qualification
```
- grouping-list 
- grouping-qualification 

分组运算,引入后,SELECT中可以出现:`共同属性`,`聚集计算结果`

此时,在分组框架下的SQL的概念执行流程

1. 计算笛卡尔积relation-list
2. 计算筛选qualification
3. 计算分组 grouping-list,按照分组属性值相等元组
4. 计算 grouping-qualification 检查,过滤
5. 计算 合格组的计算 也就是SELECT后面的`共同属性`或者`聚集计算`

> 机器的实现方式:   
> - 按照分组属性的值进行排序  
> - 再进行分组HAVING计算

共同属性必须是分组属性的子集

例,找到每一个级别最年轻水手级别,年龄需要大于18,并且得存在,得大于1
```SQL
SELECT S.rating, MIN(S.age) AS minage
FROM Sailors S
WHERE S.age > 18
GROUP BY S.rating
HAVING COUNT(*) > 1 AND EVERY (S.age <= 60)
```

> EVERY 组内每一个元素需要满足要求

**例2**: 连接基础上的 group by 统计每条红船的预定情况

```SQL
SELECT B.bid,count(*) AS count
FROM Boats B,Reserves R
WHERE R.bid=B.bid AND B.color='red'
GROUP BY B.bid
```
研究下面的SQL
```SQL
SELECT B.bid,count(*) AS count
FROM Boats B,Reserves R
WHERE R.bid=B.bid 
GROUP BY B.bid
HAVING B.color='red'
```
> 问题:`共同属性`必须是分组属性,HAVING使用的属性不是分组属性,所以这条语句应该无法通过数据库的语法检查

合格的写法:
```SQL
SELECT B.bid,count(*) AS count
FROM Boats B,Reserves R
WHERE R.bid=B.bid 
GROUP BY B.bid,B.color
HAVING B.color='red'
```

**例3**:Having里的嵌套子查询
```SQL
SELECT S.rating, MIN(S.age)
FROM Sailors S
WHERE S.age > 18
GROUP BY S.rating
HAVING 1 < (
    SELECT COUNT(*) 
    FROM Sailors S2
    WHERE S2.rating=S.rating
)
```
> **聚集函数无法嵌套**: MIN(AVG(...)),FALSE

**例子4**:FROM中的嵌套查询

```SQL
SELECT Temp.rating
FROM ( 
    SELECT S.rating,AVG(S.avg) AS avgage 
    FROM Sailors S
    GROUP BY S.rating
    ) AS Temp
WHERE Temp.avgage = (SELECT MIN(Temp.avage))
```

### CAST

### Sub-Query

子查询的三类:
- Scalar sub-qeuery 标量子查询 
- Table expression 表达式 
- Common table expression 公共表表达式

#### Scalar sub-qeuery

查询结果只是一个值,所以按照sql的文法,凡是出现一个值的地方都可以出现一个`Scalar sub-qeuery`

例1:
```SQL
SELECT d.deptname,d.location
FROM dept AS d
WHERE(
    SELECT AVG(bouns) 
    FROM emp
    WHERE deptno=d.deptno
    )>(
    SELECT AVG(salary)
    FROM emp
    WHERE deptno=d.deptno
    )
```
例2:SELECT中出现的查询
```SQL
SELECT d.deptno,d.deptname,(SELECT MAX(salary)
                   FROM emp
                   WHERE deptno=d.deptno) As maxpay
FROM dept AS d
WHERE d.location='New York'
```

#### Table expression

表达式:查询结果也是一张表

```SQL
SELECT startyear,avg(pay)
FROM( SELECT name,salary+bouns AS pay,year(startdate)as startyear 
    FROM emp) AS emp2
GROUP BY startyear;
```
在FROM子句里面在一张子句上进行查找

#### Common table expression

公共表表达式:复杂查询中表可能使用多次

`WITH`子句,相当于是临时视图(View)
```SQL
WITH payroll(deptno,totalpay) AS
   ( SELECT deptno,sum(salary)+sum(bouns)
    FROM emp
    GROUP BY deptno)
- 真正的查询开始的地方
SELECT deptno
FROM payroll
WHERE totalpay = (SELECT max(totalpay)
            FROM payroll);
```

### Outer Join

使用公共表表达式表达外连接
引入`WITH`后的SQL

```
WITH 
    table1(attr1,attr2,...) AS (
        SELECT ...
        FROM ...
        WHERE ...
        GROUP BY ...
        HAVING ...
    )
    table1(attr1,attr2,...) AS (
        SELECT ...
        FROM ...
        WHERE ...
        GROUP BY ...
        HAVING ...
    )
    ...
SELECT ...
FROM ...
WHERE ...
GROUP BY ...
HAVING ...    
```
### 递归查询

公共表自调用,递归查询增强了推理能力

```SQL
WITH agents(name,salary) AS
   ((SELECT name,salary
    FROM FedEmp
    WHERE manager='Hoover')
   UNION ALL
    (SELECT f.name,f.salary
     FROM agents AS a,FedEmp as f
     WHERE f.manager=a.name
    )
   )
SELECT name
FROM agents
```
