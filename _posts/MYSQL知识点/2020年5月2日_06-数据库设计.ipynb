## 数据之间的关系 Data Dependency
- 函数依赖(Function dependency)
- 多值依赖(Mutli-Value dependency)
- 联合依赖(Join dependency):支持无损还原的表\

## 关系模式的范式理论
- 1NF 列不可分
- 2NF 满足1NF下,不存在属性对主键的部分函数依赖
    - 插入异常
    - 删除异常
    - 更新异常
- 3NF 满足2NF下没有属性对主键的传递依赖
- (BCNF)
- 4NF 消除多值依赖
- 5NF 消除联合依赖
