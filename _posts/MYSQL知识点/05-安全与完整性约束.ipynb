## 完整性约束

> 描述数据库中关系的每一个合法示例应该满足的基本条件
> 不满足一致性要求的数据是无法进入数据库的

完整性约束规则:
- 静态约束
    - 控制数据库的静态状态(1NF)
    - 隐函约束(模式中的约束,主键约束,引用完整性约束-外键约束)
    - 显示约束 
- 动态约束:数据库在状态转换中的约束
    - 数据库状态变化时的规则

> 基于外键的约束与数据更新问题  
> 对于插入操作来说: 级联检查  
> 对于删除操作来说: 级联删除  
> 对于更新操作来说: 级联更新

现实约束的定义:
- 数据的完整性约束,在程序中指定
- 利用DBMS的断言`assert` MYSQL中的`CHECK`
- CONSTRAINT noInterLakeRes 定义一个限制
- 后面接 CHECK 表示v

```SQL
CREATE TABLE Sailors (
    sid INTEGER,
    sname CHAR(10),
    rating INTEGER,
    age REAL,
    PRIMARY KEY (sid),
    CHECK(
        rating >= 1 AND rating <= 10 
    )
)
```

内部设备禁止外借:

```SQL
CREATE TABLE Reserves (
    sname CHAR(10),
    day DATE,
    PTIMARY KEY(bid,day),
    CONSTRAINT noInterLakeRes
    CHECK(
        `Interlake` <> (
            SELECT B.bname
            FROM Boats B
            WHERE B.bid=bid 
        )
    )
)
```

## 触发器(ECA规则)

> 动态约束,数据库从一个状态到另一个状态


主动数据库系统: 能够自动监控数据库状态,并能主动采取动作(规则)

ECA规则:
- 事件: 监视ECA规则中的事件
- 条件: 触发条件
- 动作: Action

也就是触发器,也就是主动数据库系统(SQL99标准)

例子: 监视水手表,新插入的水手是否年轻,如果是,则将这个年轻水手添加到`年轻水手表`中

触发时机:
- 立即执行
- 延迟执行
- 分离执行
- 连锁触发
