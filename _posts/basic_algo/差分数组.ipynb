# 前缀和&差分

- 前缀和 -> 区间和查询
- 差分 -> 区间增减操作

对于原始数列
```
d1 d2 d3 d4 d5 ... dn
```
有差分数列,其中`fn = dn - d(n-1), f1 = d1`
```
f1 f2 f3 f4 f5 ... fn
```
有前缀和列,其中 `Sn = d1+...+dn `
```
S1 S2 S3 S4 S5 ... Sn
```
加和视图
```
# 前缀和与原始序列的关系
S1 S2 S3 S4 S5 ... Sn
----------------------
d1 d1 d1 d1 d1 ... d1
   d2 d2 d2 d2 ... d2
      d3 d3 d3 ... d3
         d4 d4 ... d4
            d5 ... d5
                ..
                 .
                   dn
                   
# 差分数组与原始列的关系
d1 d2 d3 d4 d5 ... dn
----------------------
f1 f1 f1 f1 f1 ... f1
   f2 f2 f2 f2 ... f2
      f3 f3 f3 ... f3
         f4 f4 ... f4
            f5 ... f5
                ..
                 .
                   fn
```

## 前缀和查询区间求和

sum:L->R = Sum[R] - Sum[L-1]



## 差分数组对区间操作

对区间L,R内的数字进行统一加D的操作

- d[L]   += D
- d[R+1] -= D

## 前缀序列和差分数组的关系
$S_n = \sum_{i=0}^{i=n} (n-i+1)\times f_i$

````python
    try:
        line = sys.stdin.readline()
        if line == '':
            return
        l,r = int(line.strip().split())

        l_count = int( 2*(l-1)/3 )
        r_count = int( 2*(r)/3 )
        count = r_count-l_count 
        print(count)
    except:
        pass
````

````python
import sys

line = sys.stdin.readline()
if line != '':
    l,r = int(line.strip().split())

    l_count = int( 2*(l-1)/3 )
    r_count = int( 2*(r)/3 )
    count = r_count-l_count 
    print(count)
````
