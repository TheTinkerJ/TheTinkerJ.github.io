## 日志采集

- 网页端:Aplus.JS
- 移动端:UserTrack

### 网页端采集

内容:
- 页面浏览日志采集
- 页面交互日志采集(只有你们想不到的,没有老巴做不到的)

流程:
- 由植入页面的一段JS代码来执行
- 执行时同时采集页面参数,浏览行为的上下文,运行环境信息

用户行为:
- 用户访问页面和路径
- 鼠标,焦点变化
- 页面交互反应(并不会触发浏览器加载)

日志的清洗和预处理
- 识别流量攻击,网络爬虫和流量作弊
- 数据缺项补正
- 无效数据提出
- 日志隔离分发

=> 存档在关系型数据库中

### 无线端采集

日志采集SDK对不同事件的实现

页面事件日志:
1. 设备及用户基本信息
2. 被访问页面的信息
3. 访问基本路径

UserTracker 支持在页面进入和退出时触发

在退出时触发统计可以天然统计出滞留时间

### 特殊场景

- 端内聚合
- 网页向移动端靠拢
- 浏览器 JavaScript, WebView, 客户端日志采集SDK统一

### 消息队列 TimeTunel TT

阿里巴巴使用消息队列来是实现从日志采集服务器到数据计算MawxCompute

日志采集服务器$\rightarrow MaxCompute$

- 应用可以实时订阅TT中的消息进行实时计算
- 离线应用也可以定时获取消息,完成离线计算

日志采集技术已经日臻完善,更重要的是下游的统计计算,提供符合业务特性的数据展现
