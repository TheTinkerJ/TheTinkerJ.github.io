```python
from IPython.core.display import display, HTML # 将html嵌入到python cell中
import os
indent = " " # 间隔
short = 3    # 缩进
contents = [".","_posts"]
root_path = "/".join(contents)

def show_dir(dir_name):
    left = len(contents)-short
    dir_str = [indent]*left
    dir_str.append(dir_name)
    
    print("".join(dir_str))
    
def show_note(note_name):
    left = len(contents)-short
    note_str = [indent]*left
    note_str.append(note_name)
    
    site_url = ["_site"]
    site_url.extend(contents[2:-1])
    site_url.append(note_name+".html")
    
    note_url = site_info['mat']+"/".join(site_url)
    print(note_url)
    
    aurl = '<a href="%s">%s</a>'%(note_url,note_name)
    
    display(HTML(aurl))

def search_Dir(dir_path):
    for path in os.listdir(dir_path):
        contents.append(path)
        cur_path = "/".join(contents)
#         print(cur_path)
        if os.path.isfile(cur_path) and cur_path.endswith('.ipynb'):
            filename,suffix = os.path.splitext(path)
            show_note(filename)
        elif os.path.isdir(cur_path):
            show_dir(path)
            search_Dir(cur_path)
        contents.pop()
    
search_Dir(root_path)
```

```python
def searchDir(dir_path):
    allfiles = os.listdir(dir_path)
    for file in allfiles:
        print(file)
        if file.endswith(".ipynb"):
            file_name,file_ext = os.path.splitext(file)
            print(file_name)
        elif os.path.isdir(dir_path+"/"+file):
            print("文件夹")
    
searchDir("./_posts/basic_algo")
```
