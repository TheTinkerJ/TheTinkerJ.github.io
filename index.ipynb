## 目录 Contents

```python
from IPython.core.display import display, HTML
import os

indent_html,indent_offset= "&emsp;",3        
contents = [".","_posts"]
root_path = "/".join(contents)

def show_dir(dir_name):
    left = len(contents)- indent_offset
    dir_indent = [indent_html]*left
    dir_indent.append(dir_name)
    dir_str = "".join(dir_indent) +"/"
    display(HTML(dir_str))
    
def show_note(page_name):
    left = len(contents) -indent_offset
    page_indent = [indent_html]*left
    page_indent.append(page_name)
    page_str = "".join(page_indent)
    page_url_list = ["blogs"]
    page_url_list.extend(contents[2:-1])
    page_url_list.append(page_name+".html")
    page_url = "/".join(page_url_list)
    page_url_str = '<a href="%s" style="text-decoration:none">%s</a>'%(page_url,page_str)
    display(HTML(page_url_str))

def search_Dir(dir_path):
    for path in os.listdir(dir_path):
        contents.append(path)
        cur_path = "/".join(contents)
        if os.path.isfile(cur_path) and cur_path.endswith('.ipynb'):
            filename,suffix = os.path.splitext(path)
            show_note(filename)
        elif os.path.isdir(cur_path) and not cur_path.endswith(".ipynb_checkpoints"):
            show_dir(path)
            search_Dir(cur_path)
        contents.pop()
    
search_Dir(root_path)
```
