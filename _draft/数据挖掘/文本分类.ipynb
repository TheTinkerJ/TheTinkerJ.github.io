### TF-IDF 词频-逆文档频率

找出能够表征每篇文档特征的词,具体用来评估字词对文档的重要性

$$
TF = \frac{count(w,d)}{size(d)}
$$

$$
IDF = \log\frac{n}{docs(w,D)+1}
$$

$$
TFIDF = TF*IDF 
$$

TF-IDF的使用:查询串$q$与文档$d$的的匹配程度 $tf-idf(q,d)$
