# Spark开发原则

- 避免重复创建RDD
- 尽可能复用同一个RDD
- 对多次使用的RDD进行持久化
- 尽量少使用shuffle类算子(reduceByKey,join)
- 尽量使用 map-side 预聚合的shuffle操作 combineByKey
- 使用高性能算子
    - reduceByKey/aggregateByKey->groupByKey
    - mapPartitions->map
    - filter后加一个coalesce操作
    - repartitionAndSortWithinPartitions->sort(边分区边排序)
- 大变量广播
- 使用kryo优化序列化
- 优化数据结构,尽量使用原始类型


## Job-Stage-Task

- spark是根据`shuffle类`算子进行stage划分
- 多个任务组成的并行计算阶段,因`Action`产生

> shuffle 划分 stage  
> Action 划分 Job

## Executor内存
- 20% task代码执行
- 20% shuffle下游操作
- 60% RDD持久化
