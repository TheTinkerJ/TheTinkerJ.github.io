# MYSQL 的存储引擎

MYSQL的存储引擎包含:
- MyISAM
- InnoDB

## MyISAM 对比 InnoDB


1. InnoDB支持事务,且每一条sql都默认封装成sql
2. InnoDB支持外键
3. InnoDB是聚集索引,数据索引不分家,MyISAM不同
4. InnoDB 与 MyISAM 锁类型不同,InnoDB支持行锁,但是只是逻辑实现,会退化
5. InnoDB不适用额外变量保存行数
6. InnoDB必须要有主键
7. MyISAM不支持外键

- B+树只有到达叶子节点才能命中

B树优点:

B+树优点: 中间节点不保存数据，所以磁盘页能容纳更多节点元素,必须查找到叶子节点,查找更稳定,对于范围查询来说,b+树只需遍历叶子节点链表即可,b树却需要重复地中序遍历

## 数据库索引

- B+ 树: 二分法查找
- 哈希索引: O(1)的复杂度
- 全文索引: 关键词查找

## 不用红黑树的原因

红黑树当插入删除元素的时候会进行频繁的变色与旋转，来保证红黑树的性质，浪费时间。但是当数据量较小，数据完全可以放入内存中，不需要进行磁盘IO时，红黑树时间复杂度比B+树低。


## B+树使用的原因

- 范围查询 - 层低
- 层数低 - IO次数少
- 数据只在节点稳定
