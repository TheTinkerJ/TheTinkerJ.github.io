````python
# 将奇数放在偶数前面
# 保证原始顺序不变
arr = [2,1,4,4,3,6]

p0 = 0
p1 = 0

while arr[p0] % 2 != 0:
    p0 += 1

while p1 < len(arr) and p0 < len(arr):
    while p1 < len(arr) and arr[p1] % 2 != 1:
        p1 += 1     
    if p0 < p1 < len(arr):
        tmp = arr[p1] # 奇数
        for i in range(p1,p0,-1):
            arr[i] = arr[i-1]
        arr[p0] = tmp
        p0 += 1
    p1 += 1

print(arr)
````
