# 关于jupyter scala kernel安装方法[Windows]

1. 从 github 上下载 coursier
```git
git clone https://github.com/coursier/launchers.git
```

2. 在launchers文件夹中构建almond
```
.\coursier bootstrap -r jitpack -i user -I user:sh.almond:scala-kernel-api_2.13.1:0.9.1 sh.almond:scala-kernel_2.13.1:0.9.1 ^ -o almond 
```

3. 安装almond
```
.\almond --install
```
此时已经存在scala kernel



````scala
println("hello,scala")
````

用还能用就是启动速度有点慢

参考：
- [scala kernel](https://almond.sh/docs/quick-start-install)
- [github coursier_cli](https://github.com/coursier/launchers)
