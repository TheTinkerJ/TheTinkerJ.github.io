# 我的本地sql运行环境


>  一句话描述: 安装在 window10 中 ubuntu 子系统内的 mysql


背景系统 : Windows Linux SubSystem

[参考链接](https://blog.csdn.net/weixin_43530726/article/details/91303898)

安装步骤:

```bash
# 安装mysql-server
sudo apt-get install mysql-server
# 配置权限
sudo vim /etc/mysql/mysql.conf.d/mysqld.cnf
# 在最后一行加入
skip-grant-tables
# 重启mysql服务
sudo service mysql restart
# 进入mysql
mysql -u root -p
```

````python
import pymysql 
conn = pymysql.connect(host='127.0.0.1',user='root',password='')
cursor = conn.cursor()
cursor.execute("show databases;")
cursor.fetchall()
````
