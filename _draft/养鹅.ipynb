【第一天】  
①注意温度。温度低扎堆容易压死。温度高张嘴呼吸会使体内水分迅速流失。  
②注意湿度。湿度大温度高容易滋生细菌另外湿毛。湿度低容易体内水分流失  
③保证每只鹅喝上水后再喂料。不然没饮水就吃料再喝水容易水中毒  
④等等等等。  

【第二天】  
中午温度高时候可以通风。注意不要让风直接吹到鹅。自动饮水器保证不断水。额外加一些水壶。保障每只鹅都能喝上水。如果鹅比较调皮啄毛的话加一些绳子上去给它玩。没啥事基本在躺椅上睡了一天

【第三天】  
控温器下调一度。扔点青草增加小鹅的运动量。小鹅已经熟悉环境。采食量增加。已进入生长迅速阶段。注意调高饮水器高度。小鹅。弱鹅可以顺手挑出来。集中一栏。加强管理。

【第四天】  
状态良好。该吃吃该喝喝。吃完喝完睡觉。十来分钟循环一次。趴在那里单腿伸后三三两两睡一块是最舒服的睡姿。说明棚内小环境完美。加一次料吃一天。省心很多

【第五天】  
最快乐的事就是看着小鹅慢慢长大

【第六天】  
控料了。让小鹅适当有点饥饿感

【第七天】  
等天明了要打小鹅瘟疫苗了。腿部外部肌肉注射。七号针头。打疫苗期间进行第一次分群。弱苗挑出单独一筐加强饲养
