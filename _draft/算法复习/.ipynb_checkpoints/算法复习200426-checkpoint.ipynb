## 快速排序

- 循环递归
- 右左右左右左

```python
def quickSort(arr,l,r):
    if l < r:
        mid = quickCore(arr,l,r)
        quickSort(arr,l,mid-1)
        quickSort(arr,mid+1,r)

def quickCore(arr,l,r):
    pviot = arr[l]
    while l < r:
        while l < r and arr[r] >= pviot:
            r -= 1
        arr[l] = arr[r]
        while l < r and arr[l] <= pviot:
            l += 1
        arr[r]= arr[l]
    arr[l] = pviot
    return l  

arr = [2,1,3,5,3,4]
quickSort(arr,0,len(arr)-1)
print(arr)
```

# 二叉树的遍历算法

-\[前,中,后\]$\Join$\[递归,非递归\]

```python
# 前递归
def pre_order(root):
    if root:
        print(root)
        pre_order(root.left)
        pre_order(root.right)
# 前非递归
def pre_order(root):
    opstack = []
    res = []
    if root:
        opstack.append(root)
        while len(opstack)!=0:
            cur = opstack.pop()
            res.append(cur)
            if cur.right != None:
                opstack.append(cur.right)
            if cur.left !=None:
                opstack.append(cur.left)
        return [res.pop() for _ in range(len(res))]
    else:
        return []
```

```python
# 中递归
def  mid_order(root):
    if root:
        mid_order(root.left)
        print(root)
        mid_order(root.right)
# 中非递归
def mid_order(root):
    opstack = []
    res = []
    cur = root
    while cur != None or len(opstack)!=0:
        while cur != None:
            opstack.append(cur)
            cur = cur.left
        cur = opstack.pop()
        res.append(cur)
        cur = cur.right
    return res
```
