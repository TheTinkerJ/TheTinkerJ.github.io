编辑距离的两种实现

- 递归实现
- 动态规划实现

```python
def L_distance(str1,str2):
    
    if len(str1) ==0 :
        return len(str2)
    if len(str2) == 0:
        return len(str1)
    elif str1 == str2:
        return 0
    
    if str1[len(str1)-1] == str2[len(str2)-1]:
        d = 0
    else:
        d = 1
    
    return min(
        L_distance(str1,str2[:-1]) + 1,
        L_distance(str1[:-1],str2) + 1,
        L_distance(str1[:-1],str2[:-1])+d
    )

L_distance("abc","aabcdef")
```

```python
def L_distance_dp(str1,str2):
    matrix = [[ 0 for j in range(len(str2)+1)] for i in range(len(str1)+1)]
    for i in range(1,len(str1)+1):
        for j in range(1,len(str2)+1):
            if str1[i-1] == str2[j-1]:
                d = 0
            else:
                d= 1
            matrix[i][j] = min(matrix[i-1][j]+1,matrix[i][j-1]+1,matrix[i-1][j-1]+d)
    
    for row in matrix:
        print(row)
    return matrix[-1][-1]

str1 = "abc"
str2 = "abcd"
L_distance_dp(str1,str2)
```
